```python
#importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#loading data 
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#data cleanup

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_ind= df[df['job_country'] == 'India'].copy()
df_ind['job_posted_month'] = df_ind['job_posted_date'].dt.strftime('%B')
df_ind_pivot = df_ind.pivot_table(index= 'job_posted_month', columns= 'job_title_short', aggfunc= 'size')
df_ind_pivot.reset_index(inplace =True)
df_ind_pivot['month_no'] = pd.to_datetime(df_ind_pivot['job_posted_month'], format= '%B').dt.month
df_ind_pivot.sort_values('month_no',inplace= True)
df_ind_pivot.set_index('job_posted_month', inplace = True)
df_ind_pivot.drop(columns='month_no',  inplace=True)
df_ind_pivot
```

```python
df_US_sw_pivot=pd.read_csv('https://lukeb.co/software_csv', index_col= 'job_posted_month')
df_US_sw_pivot
```

```python
df_tech_merged = df_ind_pivot.merge(df_US_sw_pivot, on= 'job_posted_month') # Similar like Inner join, merging on a common column
```

```python
top_5 = (df_tech_merged.sum()
         .sort_values(ascending= False).
         head().
         index.
         to_list())
 
top_5 
```

```python
df_tech_merged[top_5].plot(kind = 'line')
plt.title('Monthly Job posting of Tech jobs in India & US')
plt.xlabel('2023')
plt.ylabel('Job count')
plt.ylim(0,20000)
plt.legend()
plt.show()
```

```python
df_tech_merged[top_5].plot(kind = 'bar')
plt.title('Monthly Job posting of Tech jobs in India & US')
plt.xlabel('2023')
plt.ylabel('Job count')

plt.legend()
plt.show()
```
